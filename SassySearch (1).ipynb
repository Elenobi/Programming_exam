```python
import pandas as pd
import matplotlib.pyplot as plt
import re

class User:
    def __init__(self, email, password, username=None):
        self.email = email
        self.password = password
        self.username = username

    def validate_email(self):
        return re.match(r"[^@]+@[^@]+\.[^@]+", self.email)

    def validate_password(self):
        return len(self.password) >= 6 and re.search(r"[A-Z]", self.password)

class Product:
    def __init__(self, name, store, price, date_collected):
        self.name = name
        self.store = store
        self.price = price
        self.date_collected = date_collected

    def __str__(self):
        return f"{self.name} from {self.store} - ${self.price} (Date: {self.date_collected})"

class Cart:
    def __init__(self):
        self.items = []
        self.total = 0

    def add_item(self, product):
        self.items.append(product)
        self.total += product.price

    def show_cart(self):
        print("\n--- Your Cart ---")
        for idx, item in enumerate(self.items, 1):
            print(f"{idx}. {item}")
        print(f"Total: ${self.total:.2f}")

    def reset(self):
        self.items = []
        self.total = 0


def welcome_screen():
    print("Welcome to SassySearch!")
    while True:
        print("\nSelect an option:")
        print("1. Log In")
        print("2. Register")
        choice = input("Enter the number corresponding to your choice: ")

        if choice == "1":
            log_in()
            break
        elif choice == "2":
            register()
            break
        else:
            print("Invalid choice. Please enter '1' to log in or '2' to register.")

def log_in():
    print("\n--- Log In ---")
    email = input("Enter your email: ")
    password = input("Enter your password: ")

    user = User(email, password)

    if not user.validate_email():
        print("Invalid email format. Please try again.(email format:xxx@gm.com)")
        return log_in()

    if not user.validate_password():
        print("Password must be at least 6 characters and include at least one uppercase letter.")
        return log_in()

    print("Login successful! Welcome to SassySearch!", email, "How can we help you today ?")

def register():
    print("\n--- Register ---")
    username = input("Enter a username: ")
    email = input("Enter your email: ")
    password = input("Create a password: ")

    user = User(email, password, username)

    if not user.validate_email():
        print("Invalid email format. Please try again. (email format:xxx@gm.com)")
        return register()

    if not user.validate_password():
        print("Password must be at least 6 characters and include at least one uppercase letter.")
        return register()

    print(f"Registration successful! Welcome, {username}.")

def search_product(input_value, df):
    try:
        result = df[df['Product name'].str.contains(input_value, case=False, na=False) | df['Brand'].str.contains(input_value, case=False, na=False)]
        return result
    except Exception as e:
        print(f"Error during product search: {e}")
        return pd.DataFrame()

def plot_price_trends(input_value, df):
    df_filtered = search_product(input_value, df)
    if df_filtered.empty:
        print(f"No products found for '{input_value}'. Please check your input and try again.")
        return

    df_filtered = df_filtered.copy()
    df_filtered['Date collected'] = df_filtered['Date collected'].astype(str)
    df_filtered['Week'] = df_filtered['Date collected'].str.extract(r'(\d+)', expand=False).astype(int)
    df_filtered = df_filtered.sort_values('Week')

    plt.figure(figsize=(10, 6))
    for product in df_filtered['Product name'].unique():
        product_data = df_filtered[df_filtered['Product name'] == product]
        for store in product_data['Store'].unique():
            store_data = product_data[product_data['Store'] == store]
            plt.plot(store_data['Week'], store_data['Price'], label=f"{product} - {store}")

    plt.xlabel('Week (Oldest to Newest)')
    plt.ylabel('Price (USD)')
    plt.title(f'Price Trend for {input_value} Products')
    plt.xticks(df_filtered['Week'].unique())
    plt.legend()
    plt.grid(True)
    plt.show()
    print_latest_prices(input_value, df)

def print_latest_prices(input_value, df):
    df['Date collected'] = df['Date collected'].astype(str)
    df['Week'] = df['Date collected'].str.extract(r'(\d+)', expand=False).astype(int)
    df_filtered = search_product(input_value, df)
    if df_filtered.empty:
        print(f"No products found for '{input_value}'. Please check your input and try again.")
        return
    df_filtered = df_filtered.copy()
    df_filtered['Week'] = df_filtered['Date collected'].str.extract(r'(\d+)', expand=False).astype(int)
    latest_week = df['Week'].max()
    latest_week_data = df_filtered[df_filtered['Week'] == latest_week]
    if latest_week_data.empty:
        print(f"No data available for the latest week in the dataset.")
        return
    print(f"\nLatest Prices for '{input_value}' in Week {latest_week}:")
    for _, row in latest_week_data.iterrows():
        print(f"{row['Store']}: ${row['Price']} (Product: {row['Product name']})")

def payment_process():
    print("\n--- Payment Process ---")
    while True:
        print("\nSelect a payment method:")
        print("1. Card")
        print("2. Bank")
        payment_choice = input("Enter the number corresponding to your payment method: ").strip()

        try:
            if payment_choice == "1":

                card_number = input("Enter your card number (numbers only): ").strip()
                if not card_number.isdigit():
                    print("Invalid card number. Please try again.")
                    continue
                owner_name = input("Enter the card owner's name: ").strip()
                if not owner_name:
                    print("Owner name cannot be empty. Please try again.")
                    continue
                print("\nPayment successful. Thank you for your purchase!")
                break

            elif payment_choice == "2":
                bank_number = input("Enter your bank account number (numbers only): ").strip()
                if not bank_number.isdigit():
                    print("Invalid bank account number. Please try again.")
                    continue
                owner_name = input("Enter the account owner's name: ").strip()
                if not owner_name:
                    print("Owner name cannot be empty. Please try again.")
                    continue
                print("\nPayment successful. Thank you for your purchase!")
                break

            else:
                print("Invalid choice. Please select '1' for Card or '2' for Bank.")
        except ValueError as ve:
            print(f"Value Error: {ve}. Please try again.")

cart = Cart()

def proceed_purchase(df_filtered):
    global cart
    while True:
        print("\nSelect a store by entering the corresponding number:")
        stores = df_filtered['Store'].unique()
        for idx, store in enumerate(stores, 1):
            print(f"{idx}. {store}")
        try:
            store_choice = int(input("\nEnter the number corresponding to your store: "))
            if store_choice < 1 or store_choice > len(stores):
                print("Invalid store choice. Please try again.")
                continue
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue

        selected_store = stores[store_choice - 1]
        store_data = df_filtered[df_filtered['Store'] == selected_store]
        product_name = store_data.iloc[0]['Product name']
        product_price = store_data.iloc[0]['Price']
        product = Product(product_name, selected_store, product_price, store_data.iloc[0]['Date collected'])

        print("\nPrice history for product in selected store:")
        for idx, row in store_data.iterrows():
            print(f"{row['Product name']} - ${row['Price']} (Date Collected: {row['Date collected']})")
        
        while True:
            print("\nSelect an action:")
            print("1. Add to cart")
            print("2. Buy now")
            action = input("Enter the number corresponding to your action: ").strip()
            
            if action == '1':
                cart.add_item(product)
                print(f"\n'{product_name}' added to your cart. Current total: ${cart.total:.2f}")
                return  
            elif action == '2':
                cart.add_item(product)
                print(f"\n'{product_name}' added to your cart. Current total: ${cart.total:.2f}")
                print("\nDo you want to complete your order or continue shopping?")
                print("1. Complete Order")
                print("2. Continue Shopping")
                order_action = input("Enter your choice: ").strip()
                
                if order_action == "1":
                    complete_order()
                    return "complete"  
                elif order_action == "2":
                    print("\nContinuing shopping...")
                    return  
                else:
                    print("Invalid choice. Please select '1' to complete your order or '2' to continue shopping.")
            else:
                print("Invalid choice. Please enter '1' to add to cart or '2' to buy now.")

def complete_order():
    global cart
    if not cart.items:
        print("\nYour cart is empty. Add some items before completing your order!")
        return

    cart.show_cart()
    payment_process()
    print("\nThank you for your purchase!Hope to see you soon!")

    cart.reset()

df = pd.read_csv('Database_for_Sassysearch - Φύλλο1.csv')

welcome_screen() 

while True:  
    input_value = input("\nEnter a product name or brand (e.g., 'La Mer' or 'Tatcha'): ")
    print_latest_prices(input_value, df)
    plot_price_trends(input_value, df)
    df_filtered = search_product(input_value, df)
    
    if df_filtered.empty:
        print("No products found. Try searching again.")
    else:
        purchase_status = proceed_purchase(df_filtered)
        if purchase_status == "complete":
            restart_choice = input("\nPress '1' to start a new search or 'q' to quit: ").strip()
            if restart_choice == 'q':
                print("Thank you for using SassySearch! See you soon!")
                break
            elif restart_choice != '1':
                print("Invalid input. Exiting program.")
                break
```
